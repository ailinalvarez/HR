### Foundational — single table queries

1. Before the next board meeting, we want to make sure we have a clear picture of where we stand on hiring. Can you pull together a view of our current headcount situation across each team?

Easy
Skills: SELECT, WHERE, arithmetic, single-table query

2. HR has asked for a quick snapshot of our workforce composition — nothing too detailed, just a breakdown of where people currently stand in terms of their employment with us.

Easy
Skills: GROUP BY, COUNT, ORDER BY

3. The People team is preparing for their quarterly review and wants to know if there are any obvious pain points coming out of the last survey. Are there areas where employees are clearly telling us something isn't working?

Easy
Skills: JOIN (2 tables), AVG, HAVING, WHERE with subquery or filter

### Intermediate — joins & aggregation
4. Leadership wants to understand whether the engagement issues we're hearing about are company-wide or isolated to specific teams. Can you put together something that helps us compare how different departments are feeling across the main survey themes?

Intermediate
Skills: Multi-table JOIN (4+ tables), GROUP BY multiple columns, AVG, ROUND

5. One of our investors asked whether employee sentiment has been improving since we started running these surveys. We'd like to be able to give them a credible answer — can you look into this?

Intermediate
Skills: Conditional aggregation (CASE WHEN / SUM), percentage calculation, JOIN, ORDER BY date

6. We want to start recognising the employees who have consistently shown up and engaged with our feedback process. Is there a way to identify who has been with us throughout all of our surveys?

Intermediate
Skills: GROUP BY + HAVING COUNT = subquery, JOIN, filtering on aggregated counts

### Advanced — window functions, trend analysis, multi-step logic
7. A lot has changed internally over the past year and we're hoping it's showing up in how people feel. For the employees who were with us for both annual surveys, has the needle actually moved for them — and in which direction?

Intermediate+
Skills: Self-join or pivot logic, CTEs, score delta calculation, ORDER BY computed column

8. Work-life balance keeps coming up in exit interviews as a reason people leave. Before we design an intervention, can you help us understand whether this is a company-wide problem or whether certain teams have been struggling with it more persistently than others?

Advanced
Skills: Window functions (RANK / DENSE_RANK OVER PARTITION BY), CTEs, filtering on window results

9. We've been hiring a lot over the last couple of years and want to make sure the people we're bringing in are actually thriving here. Is there any evidence that newer employees are experiencing the company differently to those who have been here longer?

Advanced
Skills: Date functions, CTEs or subqueries, conditional aggregation, comparison across cohorts

10. Going into our next leadership offsite, the exec team wants a single view of engagement health across the business — something they can actually use to have an informed conversation about where to focus. Can you build that out from the latest survey data?

Advanced
Skills: Complex multi-table JOIN, conditional flags, participation rate as a derived metric, pivoted aggregation, end-to-end analytical thinking
